## Îç∞Ïù¥ÌÑ∞ Î≥ëÌï© 1Ï∞®

```python
import pandas as pd

# üîπ Îç∞Ïù¥ÌÑ∞ Î°úÎìú (ÌïÑÏöîÏãú encoding='utf-8' ÎòêÎäî 'cp949' ÏÑ§Ï†ï)
df1 = pd.read_csv('/content/drive/MyDrive/2025-Summer/·Ñâ·Ö•·ÑÄ·Öß·Üº·Ñá·Ö¢·ÑÄ·Ö™·Ñí·Ö°·Ü®·Ñå·Ö¢·ÑÉ·Ö°·Ü´/DILIRANK_SUMMARY.csv')
df2 = pd.read_excel('/content/drive/MyDrive/2025-Summer/·Ñâ·Ö•·ÑÄ·Öß·Üº·Ñá·Ö¢·ÑÄ·Ö™·Ñí·Ö°·Ü®·Ñå·Ö¢·ÑÉ·Ö°·Ü´/DILIrank-DILIscore_List (1).xlsx')

# üîπ ÎåÄÏÜåÎ¨∏Ïûê ÌÜµÏùºÏö© ÏûÑÏãú Ïó¥ ÏÉùÏÑ±
df1['Name_lower'] = df1['Synonyms'].str.lower()
df2['Name_lower'] = df2['Compound Name'].str.lower()

# üîπ ÌïÑÏöîÌïú Ïó¥Îßå ÏÑ†ÌÉùÌï¥ÏÑú Î≥ëÌï©
merged = pd.merge(df1, df2, on='Name_lower', how='left', suffixes=('_pubchem', '_liverTox'))

# üîπ Ï†ïÎ¶¨: ÏûÑÏãú Ïó¥ Ï†úÍ±∞
merged.drop(columns=['Name_lower'], inplace=True)

# üîπ Ï†ÄÏû• (option: ÏïïÏ∂ï ÎòêÎäî feather ÌòïÏãùÎèÑ Í∞ÄÎä•)
merged.to_csv('merged_result2.csv', index=False)
```
